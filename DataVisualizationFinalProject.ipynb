```python
# Import Census API Python API wrapper and Pandas Library
import censusdata
import pandas as pd

# Format Display Options
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.precision', 2)


# Pull 2018 ACS Variablse of interest from for cities which is census place equivalent. 
# The Table names were retrived from the census
# website: https://www.census.gov/programs-surveys/acs/technical-documentation/table-shells.html
# and was downloaded as csv

place = censusdata.download('acs5', 2018, censusdata.censusgeo([('place', '*')]),
                                    ['B01001_001E', 
                            
                                     "B19013_001E",
                                     'B19113A_001E','B19113B_001E','B19113C_001E',
                                     'B19113D_001E', 'B19113H_001E', 'B19113I_001E',
                                     
                                     'B19326_003E', 'B19326_006E',
                                     
                                     'B06012_001E', 'B06012_002E',
                                     
                                     'B19083_001E', 
                                     
                                     'B01002_001E',
                                     
                                     
                                     'B03002_003E',
                                     'B02001_002E', 'B02001_003E', 'B02001_004E',
                                     'B02001_005E', 'B03003_002E', 'B03003_003E',
                                    
                                    
                                    'B15003_001E','B15003_017E', 'B15003_021E', 'B15003_022E',
                                     
                                     'B08006_001E','B08006_002E','B08006_008E',
                                     
                                     'B08303_002E','B08303_003E', 'B08303_004E', 'B08303_005E',
                                     'B08303_006E', 'B08303_007E', 'B08303_008E',
                                     'B08303_009E', 'B08303_010E', 'B08303_011E', 'B08303_012E',
                                     'B08303_013E'], key)

# Re-label the Columns
place.columns = ['TotalPopulation',
                 
                 'MedianHouseholdIncome', 
                 'MedianIncomeWhiteAlone', 'MedianIncomeBlackAlone', 'MedIncomeIndianAlone',
                 'MedianIncomeAsianAlone', 'MedianIncomeNonHispanicWhite', 'MedianIncomeHispanic',
                 
                 'MedianIncomeMaleFullTime', 'MedianIncomeFemaleFullTime',
                 
                 'PovertyUniverse','BelowProverty', 
                 
                 'GiniIndex', 
                 
                 'MedianAge', 
                 
                 'NumberofWhiteOnly', 
                    'WhiteAlone', 'Black Alone', 'American Indian Alone',
                     'Asian Alone','Not Hispanic', 'Hispanic',
                'PopOver25', 'HighSchool', 'Associates', 'Bachelors',
                 
                 'WorkersOver16', 'DrivetoWork', 'PublicTransportation',
                 
                 'LessThan5', '5to9', '10to14', '15to19', '20to24', '25to29',
                 '30to34', '35to39', '40to44', '45to59', '60to89','90ormore']

place.head()
```

```python
# Sort by population in order to get the 20 most populous cities
place = place.sort_values(by=['TotalPopulation'], ascending=False)

place= place.iloc[:20]

place.loc[:,'TotalPopulation':'MedianHouseholdIncome']
```

```python
# reset the index
place = place.reset_index()
place.info()
```

```python
# Review Columns and Values
place.head()
```

```python
# Need to Reformat the index column to retrieve the City Name and State

# Split index string on ":" to drop details after city and state names
place['index'] = place['index'].astype(str)
def a (row):
    t = row['index'].split(':')
    r = t[0]
    return r


place['City']= place.apply(a, axis=1)
place.head()
```

```python
#Infor on Columns, Check that there are 20 cities and no null values
place.info()
```

```python
# Review some of the income variables
place.loc[:, ['City','PovertyUniverse','BelowProverty','GiniIndex']]
```

```python
# Create Poverty Percentage
place['PercentageBelowProverty'] = place['BelowProverty']/place['PovertyUniverse']
place.loc[:,['City', 'PercentageBelowProverty']]
```

```python
# Create Percentages for Racial Groups
place['PercentageNonHispanicWhite']= place['NumberofWhiteOnly']/place['TotalPopulation']
place['PercentageWhite']= place['WhiteAlone']/place['TotalPopulation']
place['PercentageBlack']= place['Black Alone']/place['TotalPopulation']
place['PercentageAmIndian']= place['American Indian Alone']/place['TotalPopulation']
place['PercentageAsian']= place['Asian Alone']/place['TotalPopulation']
place['PercentageHispanic']= place['Hispanic']/place['TotalPopulation']
place.loc[:, ['City', 'PercentageNonHispanicWhite','PercentageWhite', 'PercentageBlack','PercentageAmIndian',
             'PercentageAsian','PercentageHispanic' ]]
```

```python
place.info()
```

```python
# Create percentages for Education Levels
place['PercentageHighSchool'] = place['HighSchool']/place['PopOver25']
place['PercentAssociates']=place['Associates']/place['PopOver25']
place['PercentBachelors']=place['Bachelors']/place['PopOver25']
place.loc[:,['City', 'PercentageHighSchool', 'PercentAssociates', 'PercentBachelors']]
```

```python
# Create Percentages for Travel Variables
place['PercentDrivetoWork']=place['DrivetoWork']/place['WorkersOver16']
place["PercentPublic"]=place['PublicTransportation']/place['WorkersOver16']
place['TravelLessThan5']= place['LessThan5']/place['WorkersOver16']
place['Travel5to9']= place['5to9']/place['WorkersOver16']
place['Travel10to14']= place['10to14']/place['WorkersOver16']
place['Travel15to19']= place['15to19']/place['WorkersOver16']
place['Travel20to24']= place['20to24']/place['WorkersOver16']
place['Travel25to29']= place['25to29']/place['WorkersOver16']
place['Travel30to34']= place['30to34']/place['WorkersOver16']
place['Travel35to39']= place['35to39']/place['WorkersOver16']
place['Travel40to44']= place['40to44']/place['WorkersOver16']
place['Travel45to59']= place['45to59']/place['WorkersOver16']
place['Travel60to89']= place['60to89']/place['WorkersOver16']
place['Travel90ormore']= place['90ormore']/place['WorkersOver16']
place['Travel29orless'] = place['TravelLessThan5'] + place['Travel10to14'] + place['Travel15to19'] + place['Travel20to24'] +  place['Travel25to29']
place["Travel30to59"] = place['Travel30to34'] + place['Travel35to39'] + place['Travel40to44'] + place['Travel45to59']

place.loc[:,['City', 'PercentDrivetoWork',"PercentPublic",'Travel29orless', 
            'Travel30to59', 'Travel60to89', 'Travel90ormore']]
```
